```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
blank = np.zeros(img.shape,dtype="uint8")
drawImage(blank,title="Blank Image")
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImage(gray,title="Gray Image")
```

```python
blur = cv2.GaussianBlur(gray,(7,7),cv2.BORDER_DEFAULT)
drawImage(blur,title="Blurred Image")
```

```python
canny = cv2.Canny(img,125,175)
drawImage(canny,title="Canny Image")

# ret,thresh = cv2.threshold(gray,125,255,cv2.THRESH_BINARY)
# drawImage(thresh,title="Threshold Image")
```

```python
contours , hierarchy = cv2.findContours(canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
print(f'{len(contours)} countour(s) found!')
```

```python
cv2.drawContours(blank,contours,-1,(0,255,0),1)
drawImage(blank,title="Contours Drawn On Blank Image")
```
