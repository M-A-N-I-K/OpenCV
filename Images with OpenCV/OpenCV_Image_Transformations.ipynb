```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img,title="Bengal Cat")
```

```python
# Translation
def translate(img,x,y):
    transMAT = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1],img.shape[0])
    return cv2.warpAffine(img,transMAT,dimensions)
```

```python
# -x ---> Left
# -y ---> Up
# x ---> Right
# y ---> Down
translated = translate(img,-100,100)
drawImage(translated,title="Translated Image")
```

```python
# Rotation 
def rotate(img,angle,rotPoint=None):
    (height,width) = img.shape[:2]
    if rotPoint is None:
        rotPoint = (width//2,height//2)
    rot_MAT = cv2.getRotationMatrix2D(rotPoint,angle,1.0)
    dimensions = (width,height)
    return cv2.warpAffine(img,rot_MAT,dimensions)
```

```python
rotated = rotate(img,-40)
drawImage(rotated,title="Rotated Image")
```

```python
# Resizing 
resize = cv2.resize(img,(500,500),interpolation=cv2.INTER_CUBIC)
drawImage(resize,title="Resized Image")
```

```python
# Flipping
# 0 ----> Vertically flipping image
# 1 ----> Horizontally flipping image
# -1 ----> Horizontally and Vertically flipping image
flipImage = cv2.flip(img,-1)
drawImage(flipImage,title="Flipped - Image")
```

```python
# Cropping 
cropped = img[400:800,200:600]
drawImage(cropped,title="Cropped Image")
```
