```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
b,g,r = cv2.split(img)
drawImage(b,"Blue")
print(b.shape)
drawImage(g,"Green")
print(g.shape)
drawImage(r,"Red")
print(r.shape)
```

```python
merged = cv2.merge([b,g,r])
drawImage(merged,"Merged Image")
```

```python
blank = np.zeros(img.shape[:2],dtype="uint8")
blue = cv2.merge([b,blank,blank])
green = cv2.merge([blank,g,blank])
red = cv2.merge([blank,blank,r])
drawImage(blue,"Merged Blue Image")
drawImage(green,"Merged Green Image")
drawImage(red,"Merged Red Image")
```
