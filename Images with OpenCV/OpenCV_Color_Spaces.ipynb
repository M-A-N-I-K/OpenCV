```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
# BGR To Grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImage(gray,"Gray Image")
```

```python
# BGR To HSV (Hue Saturation Value)
hsv_image = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
drawImage(hsv_image,"HSV Image")
```

```python
# BGR To LAB
lab_image = cv2.cvtColor(img,cv2.COLOR_BGR2LAB)
drawImage(lab_image,"Lab Image")
```

```python
# BGR to RGB
hsv_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
drawImage(hsv_image,"RGB Image")
```
