```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
blank = np.zeros(img.shape[:2],dtype="uint8")
drawImage(blank,title="Blank Image")
```

```python
mask = cv2.rectangle(blank,(img.shape[1]//2 - 300,img.shape[0]//2 - 200),(img.shape[1]//2 + 400,img.shape[0]//2 + 200),255,-1)
drawImage(mask,title="Masked Image")
```

```python
masked_image = cv2.bitwise_and(img,img,mask=mask)
drawImage(masked_image,"Masked Image")
```
