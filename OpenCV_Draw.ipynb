# Drawing and putting text on images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
blank = np.zeros((500,500,3),dtype='uint8')
drawImage(blank)
```

```python
# Paint the image certain color
# blank[200:300,300:400] = 0,255,0
# drawImage(blank,"Green")
```

```python
# Draw a rectangle
cv2.rectangle(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2),(0,255,0),thickness=-1)
drawImage(blank)
```

```python
# Draw A Circle
cv2.circle(blank,(250,250),40,(0,0,255),thickness=-1)
drawImage(blank)
```

```python
# Draw A line
cv2.line(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2),(255,255,255),thickness=4)
drawImage(blank)
```

```python
# Write Text
cv2.putText(blank,"OpenCV",(220,250),cv2.FONT_HERSHEY_COMPLEX,0.5,(255,0,0),thickness=2)
drawImage(blank)
```
