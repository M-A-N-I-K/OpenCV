```python
import cv2
import matplotlib.pyplot as plt
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
# Converting to grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImage(gray,title="Gray Image")
```

```python
# Blurring Image
blur_image = cv2.GaussianBlur(img,(3,3),cv2.BORDER_DEFAULT)
drawImage(blur_image,title="Blur Image")
```

```python
# Edge Cascade
canny = cv2.Canny(img,125,175)
canny2 = cv2.Canny(blur_image,125,175)
drawImage(canny,title="Canny - Img")
drawImage(canny2,title="Canny - Blur Img")
```

```python
# Dilating the image
dilated = cv2.dilate(canny,(7,7),iterations=3)
drawImage(dilated,title="Dilated : Canny Image")
```

```python
# Eroding
eroded = cv2.erode(dilated,(7,7),iterations=3)
drawImage(eroded,title="Eroded : Dilated Image")
```

```python
# Resize and Cropping Image
resize = cv2.resize(img,(500,500),interpolation=cv2.INTER_CUBIC)
drawImage(resize,title="Resize Image")
```

```python
# Cropping 
cropped = img[350:1000,200:700]
drawImage(cropped,title="Cropped Image")
```
