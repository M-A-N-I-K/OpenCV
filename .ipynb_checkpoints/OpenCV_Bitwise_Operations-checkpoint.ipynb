```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
blank = np.zeros([400,400],dtype="uint8")
drawImage(blank,"Blank Image")
```

```python
rectangle = cv2.rectangle(blank.copy(),(30,30),(370,370),255,-1)
circle = cv2.circle(blank.copy(),(200,200),200,255,-1)
drawImage(rectangle,"Rectangle")
drawImage(circle,"Circle")
```

```python
# Bitwise AND
bitwise_and = cv2.bitwise_and(rectangle,circle)
drawImage(bitwise_and,"Bitwise And Image")
```

```python
# Bitwise OR
bitwise_or = cv2.bitwise_or(rectangle,circle)
drawImage(bitwise_or,"Bitwise OR Image")
```

```python
# Bitwise XOR
bitwise_xor = cv2.bitwise_xor(rectangle,circle)
drawImage(bitwise_xor,"Bitwise XOR Image")
```

```python
# Bitwise NOT
bitwise_not = cv2.bitwise_not(rectangle)
drawImage(bitwise_not,"Bitwise NOT Image")
```
