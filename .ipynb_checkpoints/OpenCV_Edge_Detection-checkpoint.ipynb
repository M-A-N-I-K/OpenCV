```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImage(gray,"Gray Image")
```

```python
# Laplacian
lap = cv2.Laplacian(gray,cv2.CV_64F)
lap = np.uint8(np.absolute(lap))
drawImage(lap,"Laplacian Image")
```

```python
# Sobel 
sobelx = cv2.Sobel(gray,cv2.CV_64F,1,0)
sobely = cv2.Sobel(gray,cv2.CV_64F,0,1)
combined_sobel = cv2.bitwise_or(sobelx,sobely)
drawImage(sobelx,"Sobel X")
drawImage(sobely,"Sobel Y")
drawImage(combined_sobel,"Combined Sobel")
```

```python
canny = cv2.Canny(gray,150,175)
drawImage(canny,"Canny Image")
```
