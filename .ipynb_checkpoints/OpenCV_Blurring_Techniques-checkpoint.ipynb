```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
# Averaging 
average = cv2.blur(img,(3,3))
drawImage(average,title="Average Blur - Image")
```

```python
# Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img,(3,3),0)
drawImage(gaussian_blur,title="Gaussian Blur - Image")
```

```python
# Median Blur
median_blur = cv2.medianBlur(img,3)
drawImage(median_blur,"Median Blur - Image")
```

```python
# Bilateral Blurring
bilateral_blur = cv2.bilateralFilter(img,5,15,15)
drawImage(img,"Bilateral Blur - Image")
```
