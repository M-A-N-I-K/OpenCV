```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImage(img,title="Image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title + str(img.shape))
    plt.show()
```

```python
img = cv2.imread("./bengal-cat-g852a4c82c_1280.jpg")
drawImage(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImage(gray,"Gray Image")
```

#### Simple Thresholding

```python
threshold , thresh = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
drawImage(thresh,"Simple Thresholded")
```

```python
threshold , thresh = cv2.threshold(gray,150,255,cv2.THRESH_BINARY_INV)
drawImage(thresh,"Simple Thresholded Inverse")
```

###### Adaptive thresholding

```python
adaptive_thresholding = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,9)
drawImage(adaptive_thresholding,"Adaptive Thresholding Image")
```
