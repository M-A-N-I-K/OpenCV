#### Capture properties

```python
# Height
height_of_frames = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
print("Height of frames :",height_of_frames)
# Width
width_of_frames = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
print("Width of frames :",width_of_frames)
```

```python
# Get Number of Frames
number_of_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print("Number of frames :",number_of_frames)
# Get Number of Frames per Second
frames_per_second = cap.get(cv2.CAP_PROP_FPS)
print("Frames per second :",frames_per_second)
# Length of video 
length = number_of_frames/frames_per_second
print("Length Of Video :",length)
```

#### Working With Video Captures

```python
# Establish Video Capture
cap = cv2.VideoCapture(os.path.join("ship.mp4"))
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    # Read Frame
    ret,frame = cap.read()
    # Gray transform
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # Show Image
    cv2.imshow("Video Player",gray)
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
# Close Down Everything
cap.release()
cv2.destroyAllWindows()
```

#### Writing out videos

```python
# Establish Video Capture
cap = cv2.VideoCapture(os.path.join("ship.mp4"))

# Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
framesPerSecond = cap.get(cv2.CAP_PROP_FPS)

# Video Writer
videoWriter = cv2.VideoWriter(os.path.join("Output","ship_gray.mp4"),cv2.VideoWriter_fourcc("M","P","G","4"),framesPerSecond,(width,height))

for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    # Read Frame
    ret,frame = cap.read()
    # Gray transform
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # Show Image
    cv2.imshow("Video Player",gray)
    # Write out frame
    videoWriter.write(gray)
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
# Close Down Everything
cap.release()
cv2.destroyAllWindows()
# Release or close video writer
videoWriter.release()
```
